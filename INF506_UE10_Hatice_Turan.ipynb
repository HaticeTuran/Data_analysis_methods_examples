```python
import seaborn as sns
import joblib 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df =  joblib.load('dfo.joblib')
df
```

```python
fig, axes = plt.subplots(2, 3, figsize=(10,6))

fig.tight_layout(pad=5.0)

axes[0,0].scatter(df.lon, df.lat, alpha = .2, s=.2)
axes[0,0].set_xlabel('Longitude (Â°)')
axes[0,0].set_ylabel('Latitude (Â°)')

axes[0,1].scatter(df.year,df.lon, alpha = .2, s=.2)
axes[0,1].set_xlabel('Year(yr)')
axes[0,1].set_ylabel('Longitude (Â°)')

axes[0,2].scatter(df.year,df.lat, alpha = .2, s=.2)
axes[0,2].set_xlabel('Year(yr)')
axes[0,2].set_ylabel('Latitude (Â°)')

axes[1,0].hist(df.lat,20, alpha = .2)
axes[1,0].set_ylabel('Frequency')
axes[1,0].set_xlabel('Latitude (Â°)')

axes[1,1].hist(df.carea,log = True , alpha = .2)
axes[1,1].set_xlabel('Carea')
axes[1,1].set_ylabel('Frequency')

axes[1,2].scatter(df.year,df.carea, alpha = .2, s=.2)
axes[1,2].set_xlabel('Year(yr)')
axes[1,2].set_ylabel('Carea')
```

```python
df['lat'].plot.density()
```

```python
fig2 = plt.figure(figsize=(8,10))
p = sns.scatterplot(data=df[ df.year <1962], x="lon",y="year", size="carea", sizes=(20,200), alpha= .3, 
                    hue = 'lat', palette='dark:salmon_r')
sns.kdeplot(data=df[ df.year <1962], x="lon",y="year", bw_adjust=.5, alpha =.9)
```
