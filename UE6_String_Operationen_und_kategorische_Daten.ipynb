```python
import pandas as pd
```

```python
df = pd.read_csv("strdata.txt", sep="\s+")
df["Full name"] = df['Name'].astype(str) +" "+  df['Surname'].astype(str)
df
```

2-Frage

```python
df["Country_code"]= df['Country'].astype("category").array.codes
df["Country_code"] = df["Country_code"].replace([0,1,2],[49,90,44])
df
```

3-Frage

```python
arr1 =[]
arr2 =[]
x=0
for i in df.Tel:
  arr1.append(str(i)[:3]) 
  arr2.append(str(i)[3:9]) 
  
df["area_code"] =arr1
df["tel_no"] =arr2
df
```

4-Frage

```python
import re

arr0 = []
arr1 = []
arr2 = []

pattern = r"([A-Z0-9._%+-]+)@([A-Z0-9.-]+)\.([A-Z]{2,4})"
regex = re.compile(pattern, flags=re.IGNORECASE)

for i in df["e-mail"]:
  
  reg = regex.match(str(i))
  arr0.append(reg.groups()[0])
  arr1.append(reg.groups()[1])
  arr2.append(reg.groups()[2])

df['username'] = arr0
df['domain'] = arr1
df['suffix'] = arr2
df
```
